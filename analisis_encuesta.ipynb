```python
# tabla de frecuencia relativa de Generation
100 * df['Generation'].value_counts() / len(df['Generation'])
```

```python
# Gráfico de tarta 
plot = df['Generation'].value_counts().plot(kind='pie', autopct='%.2f', figsize=(6, 6), title='Generation %')
```

Los resultados de esta investigación demuestran que el mayor número de encuestados esta agrupado dentro de la generación X con un porcentaje del 19.51%.

```python
# tabla de contingencia en porcentajes relativos segun género
pd.crosstab(index=df['Gender'], columns=df['Generation']).apply(lambda r: r/r.sum() *100,axis=1)
```

```python
# tabla de contingencia en porcentajes relativos segun generación
pd.crosstab(index=df['Gender'], columns=df['Gen_4']).apply(lambda r: r/r.sum() *100,axis=0)
```

```python
plot = pd.crosstab(index=df['Generation'], columns=df['Gender']).apply(lambda r: r/r.sum() *100, axis=0).plot(kind='bar')
```

De acuerdo al estudio realizado se evidenció que el género masculino tiene una proporción del 58.10% y pertenece a la generación veteranos. Sin embargo, el mayor porcentaje del género femenino, 20%, pertenecen a la generación X.

```python
pd.crosstab([df.Gen_4, df.Generation, df.Gender, df.mgr_status, df.pay_status], df.Education, margins= True)
```
